**3. Crear 1000 muestras de bootstrap con m observaciones con remplazo**

```python
import numpy as np

n = len(y)
bootstrap_coefs = []

for _ in range(1000):
    indices = np.random.choice(n, size=n, replace=True)
    x_boot = x.iloc[indices]
    y_boot = y.iloc[indices]

    coef = np.polyfit(x_boot.values.flatten(), y_boot.values.flatten(), 1)
    bootstrap_coefs.append(coef)

bootstrap_coefs = np.array(bootstrap_coefs)
```

```python
bootstrap_coefs
```

```python
# Intervalo de confianza 95%
ic_inf = np.percentile(bootstrap_coefs, 2.5, axis=0)
ic_sup = np.percentile(bootstrap_coefs, 97.5, axis=0)

print("pendiente: [{:.4e}, {:.4e}]".format(ic_inf[0], ic_sup[0]))
print("intercepto:[{:.4e}, {:.4e}]".format(ic_inf[1], ic_sup[1]))
```

## **Aggregating**

**Train**
1. Hacer train test 50% y 50%
2. Usar 3 columnas al azar
3. Filtrar matriz de diseño fit.()
4. gurdar datos de mis modelos



Hacer 1000 modelos con train

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

**Variables**

```python
X = Motor.drop(columns=['mpg', 'model'])
y = Motor['mpg']

columnas = list(X.columns)
```

**Train / test 50%**

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=42)
```

**Guardando los datos**

```python
modelos   = []   # guarda cada modelo entrenado
vars_list = []   # guarda las 3 columnas usadas por cada modelo
preds     = []   # guarda las predicciones de cada modelo en test
r2_scores_list = []   # guarda el R2 de cada modelo
```

**1000 modelos con train**

```python
from sklearn.metrics import r2_score

for i in range(1000):

    var_ = list(np.random.choice(columnas, size=3, replace=False))

    model = LinearRegression()
    model.fit(X_train[var_], y_train)

    y_pred_i = model.predict(X_test[var_])   # ŷ_test del modelo i
    r2_i     = r2_score(y_test, y_pred_i)    # R2 del modelo i

    vars_list.append(var_)
    modelos.append(model)
    preds.append(y_pred_i)
    r2_scores_list.append(r2_i)
```

**R2**

```python
r2_i = r2_score(y_test, y_pred_i)
r2_i
```

```python
y_hat_test = np.array(preds).mean(axis=0)   # ŷ_test promediado

R2_test = r2_score(y_test, y_hat_test)         # R2_test
```

**Promediar + A**

**Columnas elegidas**

```python
# Ver las 3 columnas que eligió cada modelo
for i in range(1000):
    print(f"Modelo {i+1}: {vars_list[i]}")
```

```python
from collections import Counter

conteo = Counter(col for var in vars_list for col in var)
for col, veces in conteo.most_common():
    print(f"{col}: {veces} veces")
```

**Resultados**

```python
print("ŷ_test (primeros 10):", y_hat_test[:10])
print("R2_test:", round(R2_test, 4))
```

**Conclusión**

El método de Aggregating consiste en entrenar 1,000 modelos de regresión lineal, cada uno con 3 columnas elegidas al azar de las 16 variables disponibles del dataset Motor Trend. Para cada modelo se utilizó un split de 50% train y 50% test.

Cada modelo genera un vector de predicciones ŷ de 16 observaciones, formando una matriz de 1,000x16. Al promediar las 1,000 predicciones por observación se obtiene el ŷ final agregado, el cual se compara contra los valores reales y_test para calcular el R2.

La ventaja de este enfoque es que al combinar muchos modelos entrenados con diferentes combinaciones de variables, se reduce la varianza de las predicciones y se obtiene un modelo más robusto y estable que cualquier modelo individual, ya que ninguna variable domina el resultado por sí sola.
