```python
pacientes_filtrados = Brain[(Brain['Radiation_Treatment'] == 'Yes') &
                          (Brain['Surgery_Performed'] == 'Yes') &
                          (Brain['Chemotherapy'] == 'Yes')]
pacientes_filtrados.head()
```

```python
print(f"Pacientes: {len(pacientes_filtrados)}")
```

**Escalamiento**

Definir "X" y "Y"

```python
#Tumor maligno
y = pacientes_filtrados['Tumor_Type'] == 'Malignant'
y = y.astype(int)
```

**Feature Engineering**

```python
from sklearn.preprocessing import LabelEncoder
```

```python
import pandas as pd

# pacientes
X = pacientes_filtrados.copy()
X = X.drop(columns=['Patient_ID', 'Tumor_Type'])

# Booleanos
boolean_cols = [
    'Gender', 'Radiation_Treatment', 'Surgery_Performed',
    'Chemotherapy', 'Family_History', 'MRI_Result', 'Follow_Up_Required'
]

for col in boolean_cols:
    if col == 'Gender':
        X[col] = X[col].map({'Male': 1, 'Female': 0})
    elif col == 'MRI_Result':
        X[col] = X[col].map({'Positive': 1, 'Negative': 0})
    else: # For Yes/No columns
        X[col] = X[col].map({'Yes': 1, 'No': 0})

# codigo para 'Stage' de romano a int
stage_mapping = {'I': 1, 'II': 2, 'III': 3, 'IV': 4}
X['Stage'] = X['Stage'].map(stage_mapping)

# One-Hot Encoding para columnas nom
nominal_cols = ['Location', 'Histology', 'Symptom_1', 'Symptom_2', 'Symptom_3']
X = pd.get_dummies(X, columns=nominal_cols, drop_first=True)

print("X despues de todas las transformaciones:")
display(X.head())
```

**Pipeline**

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.svm import SVC
```

```python
modelo = SVC(random_state = 42, kernel="linear", C=1)
modelo
```

```python
from sklearn.preprocessing import StandardScaler
preprocessor = StandardScaler()
pipeline = Pipeline(steps=[("preprocessor", preprocessor), ("modelo", modelo)])
```

```python
modelo.fit(X, y)
```

**Cross validation**

```python
from sklearn.model_selection import cross_val_score
cross_val_score(pipeline,X,y,scoring = "f1", cv = 5)
```

**Optimización bayesiana**

```python
from sklearn.metrics import f1_score
```

```python
from sklearn.metrics import confusion_matrix

y_pred = modelo.predict(X)

tn, fp, fn, tp = confusion_matrix(y, y_pred).ravel()

accuracy = (tp + tn) / (tp + fp + tn + fn)
recall   = tp / (tp + fn)
prec     = tp / (tp + fp)
spec     = tn / (tn + fp)
f1       = 2 * prec * recall / (prec + recall)
```

```python
accuracy, recall, prec, spec, f1
#Resultados
```
