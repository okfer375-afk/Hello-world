5. Repetir (1-4) con train-test-split 50/50

```python
#train-test-split 50/50
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=42)
```

**Regresión Lineal con Train-Test Split**

```python
# Entrenar el modelo
lr_split = LinearRegression()
lr_split.fit(X_train, y_train)
```

```python
# Calcular R^2
R_lr_split = lr_split.score(X_test, y_test)
R_lr_split
```

**Árbol con Train-Test Split**

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
# Entrenar el modelo de Árbol
dt_split = DecisionTreeRegressor()
dt_split.fit(X_train, y_train)
```

```python
# Calcular R^2
R_dt_split = dt_split.score(X_test, y_test)
R_dt_split
```
