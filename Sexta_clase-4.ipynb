## **MPG con Dummies**

```python
motor_dumi = pd.get_dummies(motor, columns=["cyl","gear","carb"], drop_first=True)
```

```python
y = motor_dumi["mpg"]
X = motor_dumi.drop(["mpg","model"], axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, train_size=0.4, random_state=24
)
```

```python
# Normalizar
sc = StandardScaler()
X_train_sc = sc.fit_transform(X_train)
X_test_sc = sc.transform(X_test)
```

```python
lr = LinearRegression()
lr.fit(X_train_sc, y_train)
```

```python
print("R2 Train:", lr.score(X_train_sc, y_train))
print("R2 Test:", lr.score(X_test_sc, y_test))
```

## **Conclusiones**

Al usar StandardScaler, estamos “emparejando el terreno” para todas las variables del modelo, para que ninguna tenga ventaja solo porque sus números son más grandes o más pequeños.

Antes de analizar todo, pusimos las variables en el mismo idioma para que el modelo las entienda de forma justa. Escalar los datos hace que el modelo trabaje de forma más ordenada, más estable y que las conclusiones que obtengamos tengan más sentido en la vida real.
